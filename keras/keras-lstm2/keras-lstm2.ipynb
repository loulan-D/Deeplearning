```python
# 用LSTM做时间序列预测
# data: 1949-1960  total 12*12 months 
# dataset:https://datamarket.com/data/set/22u3/international-airline-passengers-monthly-totals-in-thousands-jan-49-dec-60#!ds=22u3&display=line
# reference: https://machinelearningmastery.com/time-series-prediction-lstm-recurrent-neural-networks-python-keras/

# 一般的步骤：
# 1: load sequence data
# 2: standardize data
# 3: split into train data and test data
# 4: define LSTM network architecture
# 5: train LSTM network
# 6: forecast future time steps
```

```python
# import package
import numpy 
import matplotlib.pyplot as plt
from pandas import read_csv
import math
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error
# IPython 的内嵌画图，无需plt.show()
%matplotlib inline   
```

```python
# load the dataset
dataset = read_csv('./international-airline-passengers.csv',usecols = [1],engine='python',skipfooter = 3)
dataset = dataset.values
# convert int to float
dataset = dataset.astype('float32')
plt.plot(dataset)
plt.show()
```

```python
# X is the number of passengers at a given time (t) and Y is the number of passengers at the next time (t + 1).

# convert an array of values into a dataset matrix
def create_dataset(dataset,look_back = 1):
    dataX,dataY = [],[]   # len(dataX):142;len(dataY):142
    for i in range(len(dataset)-look_back-1):
        a = dataset[i:(i+look_back),0]
        dataX.append(a)
        dataY.append(dataset[i+look_back,0])
    return numpy.array(dataX),numpy.array(dataY)

# fix random  seed for reproducibility
numpy.random.seed(7)

# numpy.random.seed(7) 随机数种子，如果使用相同的seed值，每次生成的随机数都相同，如果不设置这个值，系统根据时间自己选择生成的随机数
# 则每次生成的随机数会因时间不同而不同
```

```python
# normalize the dataset
scaler = MinMaxScaler(feature_range=(0,1))
dataset = scaler.fit_transform(dataset)   # dataset.shape = (144,1)

# split into train and test data  67%train 
train_size = int(len(dataset)*0.67)
test_size = len(dataset) - train_size
train,test = dataset[0:train_size,:],dataset[train_size:len(dataset),:] # train.shape=(96,1) test.shape = (48,1)
```

```python
# x = t and y = t+1 的数据，此时维度为[samples,features]
look_back = 1
trainX,trainY = create_dataset(train,look_back) # trainX.shape = (94,1) trainY.shape = (94,)
testX,testY = create_dataset(test,look_back)    # testX.shape = (46,1)  testY.shape = (46,)

# LSTM对输入值有要求 [samples,time steps,features]
# reshape input to be [samples,time steps , features]
trainX = numpy.reshape(trainX,(trainX.shape[0],1,trainX.shape[1]))
testX = numpy.reshape(testX,(testX.shape[0],1,testX.shape[1]))

# define LSTM network architecture
# input:1 hidden:4 output:1
model = Sequential()
model.add(LSTM(12,input_shape=(1,look_back)))
model.add(Dense(1))
model.compile(loss ='mean_squared_error',optimizer = 'adam')
model.fit(trainX,trainY,epochs = 300,batch_size = 4,verbose = 2)
```

```python
# make the predictions
trainPredict = model.predict(trainX)
testPredict = model.predict(testX)
# 把预测数据转换为同一单位
trainPredict = scaler.inverse_transform(trainPredict)
trainY = scaler.inverse_transform([trainY])
testPredict = scaler.inverse_transform(testPredict)
testY = scaler.inverse_transform([testY])
trainScore = math.sqrt(mean_squared_error(trainY[0],trainPredict[:,0]))
print('train score:%.2f RMSE'%(trainScore))
testScore = math.sqrt(mean_squared_error(testY[0],testPredict[:,0]))
print('test score: %.2f RMSE'%(testScore))
```

```python
# shift train predictions for plotting
trainPredictPlot = numpy.empty_like(dataset)
trainPredictPlot[:,:] = numpy.nan
trainPredictPlot[look_back:len(trainPredict)+look_back,:] = trainPredict

# shift test predictions for plotting
testPredictPlot = numpy.empty_like(dataset)
testPredictPlot[:,:] = numpy.nan
testPredictPlot[len(trainPredict)+(look_back*2)+1:len(dataset)-1,:] = testPredict

# plot blue:before dataset; orange:trainPredictPlot;   green:testPredictPlot   
plt.plot(scaler.inverse_transform(dataset))  # before dataset
plt.plot(trainPredictPlot)
plt.plot(testPredictPlot)
```
